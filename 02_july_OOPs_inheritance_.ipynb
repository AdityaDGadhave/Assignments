1. Explain what inheritance is in object-oriented programming and why it is used.

```python
#Ans:->
#      In object-oriented programming (OOP), inheritance is a mechanism that allows a class to inherit the properties and behaviors of another
    
#      class. This is done by declaring the subclass as inheriting from the superclass. The subclass then inherits all the attributes and methods
    
#      of the superclass, and can also add its own attributes and methods.


#       The efficient code is that who not have duplication with the inheritance we can write efficient code
   
#inheritance has some Advantages As follow:->


#Reducing code duplication: Inheritance can help to reduce code duplication by allowing you to share code between classes. This can make
                          # your code more efficient and easier to maintain.

#Promoting code clarity: Inheritance can help to promote code clarity by organizing your code into a hierarchy of classes. This can make 
                        #your code easier to understand and maintain.
    
#Reusing code: Inheritance allows you to reuse code that has already been written. This can save you time and effort, and it can also help 
              #to ensure that your code is consistent.

    
#    Inheritance has diffrent type also .



class Animal:
    def __init__(self, species):
        self.species = species

    def make_sound(self):
        pass  

class Cow(Animal):                        #here Cow class is inherit with Animal class
    def __init__(self, breed):            #This type of inheritance is single inheritance
        super().__init__("Cow")
        self.breed = breed

    def make_sound(self):
        
        return ("Moo Moo")

def animal_sound(animal):
        
    return animal.make_sound()


cow = Cow("Indian")
print(cow.species)  
print(cow.breed)   
print(animal_sound(cow))  

        
Inheritance helps to organize classes in a hierarchical manner. Common functionalities shared by multiple subclasses can be grouped in the 

superclass, providing a clear and structured architecture.
```

3. Explain the terms "base class" and "derived class" in the context of inheritance.

```python
#Ans:->

#Base class:->
#             A base class is a class from which other classes inherit properties and behaviors. base class also called  superclass
            
  #           or parent class
           
   #         The base class encapsulates the common functionality that can be reused by all of its derived classes, promoting code reusability
             
  #          and a  structured class hierarchy.
            
#derived class:->
#                A derived class is a class that inherits from a base class. we can call subclass or child class for derived class   
 #   
 #               The derived class can access and use the attributes and methods of its base class. Additionally, it can override the methods
        
 #               of the base class to provide its own implementation while still retaining the rest of the inherited functionality.


class Animal:                         # Here this Animal class in base(super,parent) class
    def __init__(self, species):
        self.species = species

    def make_sound(self):
        pass  

class Cow(Animal):                              #Here Cow class is derived class(subclass or child) inherite with Animal class
    def __init__(self, breed):           
        super().__init__("Cow")
        self.breed = breed

    def make_sound(self):
        
        return ("Moo Moo")

def animal_sound(animal):
        
    return animal.make_sound()


cow = Cow("Indian")
print(cow.species)  
print(cow.breed)   
print(animal_sound(cow))  





The terms "base class" and "derived class" are often used interchangeably, but there is a subtle difference between the two terms.

The base class is the class that is being inherited from, while the derived class is the class that is doing the inheriting.
```

4. What is the significance of the "protected" access modifier in inheritance? How does
it differ from "private" and "public" modifiers?

```python
#Ans:->
 #     In object-oriented programming (OOP), the protected access modifier is used to control the accessibility of members of a class. Members
    
 #     that are declared as protected are accessible to
        
 #    The class in which they are declared
    
#     Derived classes of the class in which they are declared
        
        
        
#Protected members are not accessible to other classes. This means that they cannot be accessed directly by objects of other classes. 

# derived classes can access protected members, which allows them to reuse the code and functionality of the base class.


#The private access modifier is used to make members of a class inaccessible to other classes. This means that private members can only be 

#accessed by the class in which they are declared.

#The public access modifier is used to make members of a class accessible to all classes. This means that public members can be accessed by

#objects of any class.


#The protected access modifier is a good choice for members that should be accessible to derived classes, but not to other classes. 

#This is because it allows derived classes to reuse the code and functionality of the base class, while still preventing other classes 

#from accessing the members directly.



#Private:->	 Only accessible to the class in which it is declared.

#Private Access Modifier:-

class MyClass:
    def __init__(self):
        self._private = 78

    def _private_method(self):
        return "Private class"

obj = MyClass()
print(obj._private)       
print(obj._private_method())



#Protected:-> Accessible to the class in which it is declared, and to derived classes of the class in which it is declared.

class MyBaseClass:
    def __init__(self):
        self._protected_var = 85

    def _protected_method(self):
        return "This is protected"

class MyDerivedClass(MyBaseClass):
    def __init__(self):
        super().__init__()

    def access_protected_member(self):
        print(self._protected_var)       
        print(self._protected_method())  

obj = MyDerivedClass()
obj.access_protected_member()


The "protected" access modifier provides a balance between the openness of "public" and the encapsulation of "private."

The "protected" access modifier provides a balance between the openness of "public" and the encapsulation of "private."



#Public:->	 Accessible to all classes.

#Public Access Modifier:->

class MyClass:
    def __init__(self):
        self.public = 52

obj = MyClass()
print(obj.public)
```

6. Create a base class called "Vehicle" with attributes like "make", "model", and "year".
Then, create a derived class called "Car" that inherits from "Vehicle" and adds an
attribute called "fuel_type". Implement appropriate methods in both classes.

```python
#Ans:->

class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def print_details(self):
        print("Make:", self.make)
        print("Model:", self.model)
        print("Year:", self.year)

class Car(Vehicle):
    def __init__(self, make, model, year, fuel_type):
        super().__init__(make, model, year)
        self.fuel_type = fuel_type

    def print_details(self):
        super().print_details()
        print("Fuel type:", self.fuel_type)

car = Car("Tata", "Harrier", 2023, "Disel")
car.print_details()



This code first defines the Vehicle class with three attributes: make, model, and year. Then, it defines the Car class as a subclass of 

Vehicle. The Car class adds an additional attribute called fuel_type.

The code then creates a Car object and prints the details of the car. The print_details() method is overridden in the Car class to print 

the fuel type of the car.
```

attribute called "department" for the "Manager" class and "programming_language"

8. Design a base class called "Shape" with attributes like "colour" and "border_width."
Create derived classes, "Rectangle" and "Circle," that inherit from "Shape" and add
specific attributes like "length" and "width" for the "Rectangle" class and "radius" for
the "Circle" class.

```python
#Ans:->

class Shape:
    def __init__(self, colour, border_width):
        self.colour = colour
        self.border_width = border_width

    def draw(self):
        print("Drawing a shape with colour {} and border width {}".format(
            self.colour, self.border_width))

class Rectangle(Shape):
    def __init__(self, colour, border_width, length, width):
        super().__init__(colour, border_width)
        self.length = length
        self.width = width

    def draw(self):
        super().draw()
        print("Rectangle with length {} and width {}".format(self.length, self.width))

class Circle(Shape):
    def __init__(self, colour, border_width, radius):
        super().__init__(colour, border_width)
        self.radius = radius

    def draw(self):
        super().draw()
        print("Circle with radius {}".format(self.radius))

def main():
    rectangle = Rectangle("Saffron", 2, 10, 5)
    rectangle.draw()

    circle = Circle("red", 1, 10)
    circle.draw()


main()

In this example, we have a base class "Shape" with attributes "colour" and "border_width." The "Rectangle" class inherits from "Shape" and

adds two specific attributes, "length" and "width." The "Circle" class also inherits from "Shape" and adds a specific attribute "radius."
```

10. Create a base class called "BankAccount" with attributes like "account_number" and
"balance." Derive two classes, "SavingsAccount" and "CheckingAccount," from
"BankAccount." Add specific methods like "calculate_interest" for the
"SavingsAccount" class and "deduct_fees" for the "CheckingAccount" class.

```python
#Ans:->

class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance


class SavingsAccount(BankAccount):
    def __init__(self, account_number, balance, interest_rate):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate

    def calculate_interest(self):
        interest_amount = self.balance * (self.interest_rate / 100)
        self.balance += interest_amount
        return interest_amount


class CheckingAccount(BankAccount):
    def __init__(self, account_number, balance, fee):
        super().__init__(account_number, balance)
        self.fee = fee

    def deduct_fees(self):
        if self.balance >= self.fee:
            self.balance -= self.fee
            return self.fee
        else:
            return "Insufficient balance to deduct fees."



savings_account1 = SavingsAccount("65874615", 7000, 2.5)
checking_account1 = CheckingAccount("22554155", 5000, 50)

print("Savings Account:")
print("Account Number:", savings_account1.account_number)
print("Balance:", savings_account1.balance)
print("Interest Rate:", savings_account1.interest_rate, "%")
print("Interest Amount:", savings_account1.calculate_interest())
print("Updated Balance:", savings_account1.balance)

print("\nChecking Account:")
print("Account Number:", checking_account1.account_number)
print("Balance:", checking_account1.balance)
print("Fee:", checking_account1.fee)
print("Fee Deducted:", checking_account1.deduct_fees())
print("Updated Balance:", checking_account1.balance)

BankAccount: The base class. It has two attributes: account_number and balance. It also has two methods: deposit() and withdraw().

SavingsAccount: The derived class of BankAccount. It adds one attribute: interest_rate. It also overrides the calculate_interest() method
to calculate the interest earned on the savings account balance.
CheckingAccount: The derived class of BankAccount. It adds one attribute: monthly_fee. It also overrides the deduct_fees() method to deduct the monthly fee from the checking account balance.
The main() function creates two objects, a SavingsAccount object and a CheckingAccount object. It then calls the calculate_interest() method on the SavingsAccount object and the deduct_fees() method on the CheckingAccount object.
```
