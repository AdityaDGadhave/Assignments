3. What is a class in Python?

```python
#Ans:->
#      A class in Python is a blueprint for creating objects. It is a way to organize data and behavior into a single unit. Classes can 
    
#      be used to represent real-world entities, such as people, cars, and houses.
        
#      It defines a logical structure that specifies how the objects of that class will behave and what attributes they will have  


#      In python we define class using the keyword 'class' 

class Country:
    def __init__(self,country_name,country_code):
        self.country_name=country_name
        self.country_code=country_code
        
    def info(self):
        print(f'My country is { self.country_name} and code is {self.country_code}')
        
        
my_object=Country('India',+91)        

my_object.info()
    
    
#The Country class is a blueprint for creating Country objects. The __init__() method is called when a new Country object is created, and it 

#is used to initialize the object's attributes. The info method is a method that performs an action, in this case, it prints a all 

#information

```

7. For a library management system, you have to design the "Book" class with OOP
principles in mind. The “Book” class will have following attributes:
a. title: Represents the title of the book.
b. author: Represents the author(s) of the book.
c. isbn: Represents the ISBN (International Standard Book Number) of the book.
d. publication_year: Represents the year of publication of the book.
e. available_copies: Represents the number of copies available for checkout.
The class will also include the following methods:
a. check_out(self): Decrements the available copies by one if there are copies
available for checkout.
b. return_book(self): Increments the available copies by one when a book is
returned.
c. display_book_info(self): Displays the information about the book, including its
attributes and the number of available copies.

```python
#Ans:->

class Book:

    def __init__(self, title, author, isbn, publication_year, available_copies):  #class had 5 Attribute
        self.title = title
        self.author = author
        self.isbn = isbn
        self.publication_year = publication_year
        self.available_copies = available_copies

    def check_out(self):                        #this method for checkout
        if self.available_copies > 0:
            self.available_copies -= 1

    def return_book(self):                     #For return book
        self.available_copies += 1
 
    def display_book_info(self):              #All information of book
        print("Title:", self.title)
        print("Author:", self.author)
        print("ISBN:", self.isbn)
        print("Publication year:", self.publication_year)
        print("Available copies:", self.available_copies)


def main():
    book = Book("Full stack Data science", "Krish Naik", "0345391838", 2023, 5)   #object created inside the fuction
    print("Checking out a book...")
    book.check_out()
    print("Displaying book information...")
    book.display_book_info()
    print("Returning a book...")
    book.return_book()
    print("Displaying book information...")
    book.display_book_info()



main()



In this implementation, the Book class has attributes for the title, author, ISBN, publication year, and available copies of the book. The

class includes methods check_out() and return_book() for managing the available copies when books are checked out or returned. The 

display_book_info() method is provided to print the information about the book, including its attributes and the number of available

copies.
```

8. For a ticket booking system, you have to design the "Ticket" class with OOP
principles in mind. The “Ticket” class should have the following attributes:
a. ticket_id: Represents the unique identifier for the ticket.
b. event_name: Represents the name of the event.
c. event_date: Represents the date of the event.
d. venue: Represents the venue of the event.
e. seat_number: Represents the seat number associated with the ticket.
f. price: Represents the price of the ticket.
g. is_reserved: Represents the reservation status of the ticket.
The class also includes the following methods:
a. reserve_ticket(self): Marks the ticket as reserved if it is not already reserved.
b. cancel_reservation(self): Cancels the reservation of the ticket if it is already
reserved.
c. display_ticket_info(self): Displays the information about the ticket, including its
attributes and reservation status.

```python
#Ans:->

class Ticket:
    """This class represents a ticket in a ticket booking system."""

    def __init__(self, ticket_id, event_name, event_date, venue, seat_number, price, is_reserved):
        """This is the __init__ method. It is called when a new object is created."""
        self.ticket_id = ticket_id
        self.event_name = event_name
        self.event_date = event_date
        self.venue = venue
        self.seat_number = seat_number
        self.price = price
        self.is_reserved = is_reserved

    def reserve_ticket(self):        # This method marks the ticket as reserved if it is not already reserved.
       
        if not self.is_reserved:
            self.is_reserved = True

    def cancel_reservation(self):
                                        #This method cancels the reservation of the ticket if it is already
                                          #   reserved.
        if self.is_reserved:
            self.is_reserved = False

    def display_ticket_info(self):
                                               #This method displays the information about the ticket, including its
                                                # attributes and reservation status.
        print("Ticket ID:", self.ticket_id)
        print("Event name:", self.event_name)
        print("Event date:", self.event_date)
        print("Venue:", self.venue)
        print("Seat number:", self.seat_number)
        print("Price:", self.price)
        print("Reservation status:", self.is_reserved)



ticket = Ticket(20230714, "Isro_chandrayan_3 Launch ", "2023-07-14", "Shriharikotta lauch side", 121, 100, False)
print("Ticket information...")
ticket.display_ticket_info()
print("Reserving ticket...")
ticket.reserve_ticket()
print("Ticket information again...")
ticket.display_ticket_info()
print("Cancelling reservation...")
ticket.cancel_reservation()
print("Ticket information again...")
ticket.display_ticket_info()

#In this implementation, the Ticket class has attributes for ticket_id, event_name, event_date, venue, seat_number, price, and is_reserved.

#The class includes methods reserve_ticket() and cancel_reservation() for managing the reservation status of the ticket. 

#The display_ticket_info() method is provided to print the information about the ticket, including its attributes and reservation status.
```

9. You are creating a shopping cart for an e-commerce website. Using OOP to model
the "ShoppingCart" functionality the class should contain following attributes and
methods:
a. items: Represents the list of items in the shopping cart.
The class also includes the following methods:

a. add_item(self, item): Adds an item to the shopping cart by appending it to the
list of items.
b. remove_item(self, item): Removes an item from the shopping cart if it exists in
the list.
c. view_cart(self): Displays the items currently present in the shopping cart.
d. clear_cart(self): Clears all items from the shopping cart by reassigning an
empty list to the items attribute.

```python
#Ans:->

class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)
        print(f"{item} added to the shopping cart.")

    def remove_item(self, item):
        if item in self.items:
            self.items.remove(item)
            print(f"{item} removed from the shopping cart.")
        else:
            print(f"{item} is not present in the shopping cart.")

    def view_cart(self):
        if not self.items:
            print("The shopping cart is empty.")
        else:
            print("Items in the shopping cart:")
            for item in self.items:
                print(f" - {item}")

    def clear_cart(self):
        self.items = []
        print("Shopping cart cleared.")


cart = ShoppingCart()

cart.add_item("Laptop")
cart.add_item("Phone")
cart.add_item("Playstation")

cart.view_cart()

cart.remove_item("Phone")
cart.remove_item("Tablet")  

cart.view_cart()

cart.clear_cart()
cart.view_cart()
  
    
#In this implementation, the ShoppingCart class has an attribute items, which is a list representing the items present in the shopping cart. 

#The class includes methods add_item(), remove_item(), view_cart(), and clear_cart() to manage the shopping cart's functionality.
```

10. Imagine a school management system. You have to design the "Student" class using
OOP concepts.The “Student” class has the following attributes:
a. name: Represents the name of the student.
b. age: Represents the age of the student.
c. grade: Represents the grade or class of the student.
d. student_id: Represents the unique identifier for the student.
e. attendance: Represents the attendance record of the student.
The class should also include the following methods:
a. update_attendance(self, date, status): Updates the attendance record of the
student for a given date with the provided status (e.g., present or absent).
b. get_attendance(self): Returns the attendance record of the student.
c. get_average_attendance(self): Calculates and returns the average
attendance percentage of the student based on their attendance record.

```python
#Ans:->
class Student:
 

    def __init__(self, name, age, grade, student_id): 
         
        self.name = name
        self.age = age
        self.grade = grade
        self.student_id = student_id
        self.attendance = []

    def update_attendance(self, date, status):          #This method for update the attendence
        
        self.attendance.append({
            "date": date,
            "status": status
        })

    def get_attendance(self):          #Returns the attendance record of the student.
        
        return self.attendance

    def get_average_attendance(self):                   #For calculating average
        
        total_days = len(self.attendance)
        present_days = sum(attendance["status"] == "present" for attendance in self.attendance)
        average_attendance = (present_days / total_days) * 100
        return average_attendance



student = Student("Aditya Gadhaev", 23, "Graduation", 129610)
student.update_attendance("2023-07-25", "present")
student.update_attendance("2023-07-26", "absent")
student.update_attendance("2023-07-27", "present")
print("Attendance record:")
print(student.get_attendance())
print("Average attendance:", student.get_average_attendance())
```
